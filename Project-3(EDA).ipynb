Dataset contains cases from study condected between 1958 to 1970 at Chicago's Billings Hospital on survival of patients who had undergo surgery for BREAST CANCER

->A Dataset of Breast Cancer paticents who survivied or died after operation within 5 years.

->Instances : 306

->Attributes :4(Includind the class Attribute).

->At time of operation:-
 
 a.Age of paticent.

 b.Paticent year of operation.
 
 c.Number of Positive axillary nodes detected.
 
 d.Survival status(class attribute).
   >if 1=paticent survived till 5 years or longer.
   
   >if 2=patcient died within 5 years.

# CDF 

```python
data_1=data.loc[data["status"]==1]
data_2=data.loc[data["status"]==2]
```

```python
counts, bin_edges = np.histogram(data_1["age"], bins=10,density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], cdf)

plt.show();
```

```python
#Observed:- in 5 ys mostly survived people are below age 70 yrs
#And mostly 50 to 60 yrs had mostyl survived in thes 5 yrs
```

```python
counts, bin_edges = np.histogram(data_2['age'], bins=10,density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], cdf)
plt.legend()

plt.show();
```

```python
#conclustion: it is observed that bove 70% of people had died with in these 5 yrs ismosty equal to 60 yrs
```

# MEAN AND STANDARD DEVIATION

```python
print("Means:")
print(np.mean(data_1["year"]))
#Mean with an outlier.
print(np.mean(np.append(data_1["year"],30)));
print(np.mean(data["year"]))

print("\nStd-dev:");
print(np.std(data_1["year"]))
print(np.std(data_2["year"]))
```

# Median,Quantiles,Percentiles,IQR

```python
#Median, Quantiles, Percentiles, IQR.
print("\nMedians:")
print(np.median(data_1["age"]))
#Median with an outlier
print(np.median(np.append(data_1["age"],50)));
print(np.median(data_2["age"]))


print("\nQuantiles:")
print(np.percentile(data_1["age"],np.arange(0, 100, 25)))
print(np.percentile(data_2["age"],np.arange(0, 100, 25)))

print("\n90th Percentiles:")
print(np.percentile(data_1["age"],90))
print(np.percentile(data_2["age"],90))

from statsmodels import robust
print ("\nMedian Absolute Deviation")
print(robust.mad(data_1["age"]))
print(robust.mad(data_2["age"]))
```
